```python
"""
titanic_example.py
==================
Exemplo de uso do AutoClassificationEngine com o dataset Titanic (raw CSV).

Fluxo:
  1. Carrega e cria features derivadas (title, family_size, is_alone)
  2. Faz split estratificado treino/teste
  3. Treina o engine
  4. Exibe leaderboard e relat√≥rio visual completo
  5. Salva o bundle e demonstra o carregamento

Compat√≠vel com: auto_classification_engine.py (vers√£o com Theil's U / Eta¬≤)
"""

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Importa o engine do mesmo diret√≥rio (ajuste o path se necess√°rio)
from src.meu_framework import AutoClassificationEngine
```

```python
# ---------------------------------------------------------------------------
# 1. CARGA E FEATURE ENGINEERING
# ---------------------------------------------------------------------------


def load_and_prep_titanic() -> pd.DataFrame:
    print("üö¢ Carregando Dataset Titanic (Raw CSV)...")
    url = (
        "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
    )
    df = pd.read_csv(url)
    df.columns = [c.lower() for c in df.columns]

    # Garante que colunas objeto sejam string (evita mistura com NaN nativo)
    for col in df.select_dtypes(["object", "category"]).columns:
        df[col] = df[col].astype(str)

    print("‚ú® Criando Features Derivadas...")

    # Extrai t√≠tulo do nome antes de descartar a coluna
    df["title"] = df["name"].str.extract(r" ([A-Za-z]+)\.", expand=False)
    rare_titles = [
        "Lady",
        "Countess",
        "Capt",
        "Col",
        "Don",
        "Dr",
        "Major",
        "Rev",
        "Sir",
        "Jonkheer",
        "Dona",
    ]
    df["title"] = df["title"].replace(rare_titles, "Rare")
    df["title"] = df["title"].replace({"Mlle": "Miss", "Ms": "Miss", "Mme": "Mrs"})

    # Features de fam√≠lia
    df["family_size"] = df["sibsp"] + df["parch"] + 1
    df["is_alone"] = (df["family_size"] == 1).astype(int)

    # 'name' foi fonte do 'title'; agora √© apenas texto livre de alta cardinalidade.
    # Adicionamos aqui para n√£o depender s√≥ do sanity_check autom√°tico.
    df = df.drop(columns=["name"], errors="ignore")

    print(f"   Dataset: {df.shape[0]} linhas √ó {df.shape[1]} colunas")
    return df


# ---------------------------------------------------------------------------
# 2. SPLIT ESTRATIFICADO
# ---------------------------------------------------------------------------

df = load_and_prep_titanic()

train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df["survived"],  # mant√©m propor√ß√£o de sobreviventes em ambos os splits
)

print(f"\nüìÇ Treino: {train_df.shape[0]} linhas | Teste: {test_df.shape[0]} linhas")
print(
    f"   Taxa de sobreviv√™ncia ‚Äî Treino: {train_df['survived'].mean():.1%} | "
    f"Teste: {test_df['survived'].mean():.1%}"
)


# ---------------------------------------------------------------------------
# 3. CONFIGURA√á√ÉO DO ENGINE
# ---------------------------------------------------------------------------

params = {
    "target": "survived",
    # passengerid ‚Üí ID puro
    # ticket     ‚Üí alta cardinalidade sem valor preditivo direto
    # cabin      ‚Üí >77% nulo; o que existe tem cardinalidade alt√≠ssima
    # name       ‚Üí j√° dropado na carga ap√≥s extra√ß√£o do title
    "features_to_exclude": ["passengerid", "ticket", "cabin", "is_alone"],
    "pipeline_settings": {
        "use_pca": False,
        "pca_components": 0.95,
        "corr_threshold": 0.80,  # threshold para remo√ß√£o de colinearidade
    },
    # For√ßa tipos categ√≥ricos para vari√°veis que o pandas leria como int/float
    "force_types": {
        "category": ["sex", "embarked", "pclass", "title"],
    },
    # average_precision √© mais informativa que AUC-ROC em datasets desbalanceados
    "eval_metric": "roc_auc",
    # high_quality: bom balan√ßo entre capacidade e risco de overfitting
    # em datasets pequenos como o Titanic (~700 linhas de treino)
    "presets": "high_quality",
    # time_limit aqui serve de fallback; pode ser sobrescrito no fit()
    "time_limit": 60,
    "use_importance_filter": True,  # Ativa a Etapa 3 (corta features in√∫teis)
}


# ---------------------------------------------------------------------------
# 4. TREINAMENTO
# ---------------------------------------------------------------------------

print("\n‚öôÔ∏è  Inicializando Engine...")
engine = AutoClassificationEngine(params)

print("\nüèãÔ∏è  Treinando...")
leaderboard = engine.fit(train_df)  # usa time_limit=60 do params

print("\nüèÜ Top 5 Modelos:")
print(
    leaderboard[["model", "score_val", "pred_time_val"]].head(5).to_string(index=False)
)


# ---------------------------------------------------------------------------
# 5. CROSS-VALIDATION
# ---------------------------------------------------------------------------

print("\nüîÑ Executando Cross-Validation (5 folds estratificados)...")
# time_limit_per_fold=60 ‚Üí cada fold treina por at√© 60s
# Total estimado: ~5min (pode ajustar conforme tempo dispon√≠vel)
cv_summary = engine.cross_validate(train_df, n_folds=5, time_limit_per_fold=60)

print("\nüìä Resumo CV:")
print(cv_summary.to_string())

print("\nüìà Plotando Relat√≥rio de CV...")
engine.plot_cv_report()


# ---------------------------------------------------------------------------
# 6. RELAT√ìRIO DE ASSOCIA√á√ïES
# ---------------------------------------------------------------------------

print("\nüîó Plotando Relat√≥rio de Associa√ß√µes (Pearson / Theil's U / Eta¬≤)...")
engine.plot_association_report()


# ---------------------------------------------------------------------------
# 6. RELAT√ìRIO DE PERFORMANCE
# ---------------------------------------------------------------------------

print("\nüìä Gerando Relat√≥rio Visual de Performance...")
# bins=5 (Quintis): base de teste pequena (~180 obs) ‚Üí 5 faixas d√£o ordena√ß√£o
# mais est√°vel do que 10 decis que podem ficar ruidosos.
engine.plot_complete_report(test_df, train_data=train_df, bins=5)


# ---------------------------------------------------------------------------
# 7. INFER√äNCIA COM A NOVA API P√öBLICA
# ---------------------------------------------------------------------------

print("\nüîÆ Testando API de infer√™ncia p√∫blica...")
probs = engine.predict_proba(test_df)
classes = engine.predict(test_df)

pos_label = engine.predictor.positive_class
print(
    f"   Probabilidade m√©dia de sobreviver (pos_label={pos_label}): "
    f"{probs[pos_label].mean():.3f}"
)
print(f"   Predi√ß√µes (primeiras 10): {classes.values[:10].tolist()}")


# ---------------------------------------------------------------------------
# 8. SAVE + LOAD
# ---------------------------------------------------------------------------

# BUNDLE_PATH = "modelo_titanic_prod"

# engine.save_bundle(BUNDLE_PATH)

# print(f"\nüìÇ Testando carregamento do bundle salvo em '{BUNDLE_PATH}'...")
# engine_loaded = AutoClassificationEngine.load(BUNDLE_PATH)

# # Sanity check: predi√ß√µes devem ser id√™nticas
# probs_loaded = engine_loaded.predict_proba(test_df)
# diff = (probs[pos_label] - probs_loaded[pos_label]).abs().max()
# print(f"   Diferen√ßa m√°xima de probabilidade entre original e loaded: {diff:.2e}")
# assert diff < 1e-6, "‚ö†Ô∏è ATEN√á√ÉO: Predi√ß√µes divergem ap√≥s load!"
# print("   ‚úÖ Engine recarregado com sucesso ‚Äî predi√ß√µes id√™nticas.\n")

# print("‚úÖ Processo conclu√≠do!")
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

# 1. Carrega o dataset
print("‚è≥ Fazendo download do dataset German Credit (OpenML)...")
data = fetch_openml(name="credit-g", version=1, as_frame=True, parser="auto")
df = data.frame.copy()

# ---------------------------------------------------------------------------
# üõ† ENGENHARIA DE VARI√ÅVEIS (FORA DO FRAMEWORK)
# ---------------------------------------------------------------------------
print("üîß Criando novas features estrat√©gicas...")

# A. Parcela Mensal Estimada: Valor do cr√©dito dividido pela dura√ß√£o
# Ajuda o modelo a entender o peso do compromisso mensal.
df["installment_per_month"] = df["credit_amount"] / df["duration"]

# B. Intera√ß√£o de Risco Et√°rio: Jovens com pouco tempo de emprego costumam ser mais inst√°veis.
# Criamos uma flag para menores de 25 anos com menos de 1 ano de emprego.
df["is_young_unstable"] = (
    (df["age"] < 25) & (df["employment"].isin(["<1", "unemployed"]))
).astype(int)

# C. Score Consolidado de Reservas:
# For√ßamos o .astype(float) para o Pandas entender que agora queremos fazer conta!
mapping_status = {"no checking": 0, "<0": 1, "0<=X<200": 2, ">=200": 3}
mapping_savings = {
    "no known savings": 0,
    "<100": 1,
    "100<=X<500": 2,
    "500<=X<1000": 3,
    ">=1000": 4,
}

df["internal_reserve_score"] = df["checking_status"].map(mapping_status).astype(
    float
).fillna(0) + df["savings_status"].map(mapping_savings).astype(float).fillna(0)

# D. SIMULA√á√ÉO DE DADO EXTERNO (O 'BUREAU SCORE')
# Adicionamos o .astype(float) logo ap√≥s o map para permitir a soma com o numpy
np.random.seed(42)
df["external_bureau_score"] = df["class"].map({"good": 700, "bad": 300}).astype(
    float
) + np.random.normal(0, 150, len(df))

# ---------------------------------------------------------------------------

# 2. Split (Usando o DF modificado)
df_treino, df_teste = train_test_split(
    df, test_size=0.20, stratify=df["class"], random_state=42
)

print(f"Treino: {df_treino.shape} | Teste: {df_teste.shape}")

# 3. Configura o Modo Turbo
params_credit = {
    "target": "class",
    "eval_metric": "roc_auc",
    "time_limit": 180,  # 3 minutos costumam ser suficientes com estas features
    "presets": "high_quality",
    "use_importance_filter": True,
    "use_sklearn_pipeline": False,
    "prune_models": True,
    "num_cpus": 6,
    "ag_args_ensemble": {"fold_fitting_strategy": "sequential_local"},
    "pipeline_settings": {"corr_threshold": 0.88},
}

# 4. Treina
engine_credit = AutoClassificationEngine(params_credit)
engine_credit.fit(df_treino)

# ---------------------------------------------------------------------------
# 6. RELAT√ìRIOS
# ---------------------------------------------------------------------------
print("\nüîó Relat√≥rio de Associa√ß√µes...")
engine_credit.plot_association_report()

print("\nüìä Relat√≥rio de Performance...")
engine_credit.plot_complete_report(test_data=df_teste, train_data=df_treino, bins=5)

# ---------------------------------------------------------------------------
# 8. INFER√äNCIA
# ---------------------------------------------------------------------------
print("\nüîÆ Testando API de infer√™ncia...")
probs = engine_credit.predict_proba(df_teste)
pos_label = engine_credit.predictor.positive_class
print(f"   Probabilidade m√©dia '{pos_label}': {probs[pos_label].mean():.3f}")
```
